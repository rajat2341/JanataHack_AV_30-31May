```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings("ignore")
```

```python
import re
```

```python
train = pd.read_csv("data/train_JH_AV.csv")
test = pd.read_csv("data/test_JH_AV.csv")
```

```python
train['label'] = "train"
test['label'] = "test"
```

```python
final = pd.concat([train, test], sort = False)
```

## Data Transformation 

```python
def clean(data):
    cleanr = re.compile(',')
    cleaned = re.sub(cleanr, '', data)
    return int(cleaned)

temp = final['Loan_Amount_Requested']
temp1 = temp.map(clean)
final['Loan_Amount_Requested'] = temp1
```

```python
final['Income_Verified'].replace(('VERIFIED - income source'), ('VERIFIED - income'), inplace = True)
```

```python
final['Length_Employed'].unique()
```

```python
final['Length_Employed'].replace(('< 1 year', '4 years', '7 years', '8 years', '2 years',
       '10+ years', '1 year', '6 years', '9 years', '3 years',
       '5 years'), (0,4,7,8,2,10,1,6,9,3,5), inplace = True)
```

```python
final['Home_Owner'].unique()
```

```python
final['Home_Owner'].replace(('Rent', 'Mortgage', 'Own', 'Other', 'None'), (1, 2, 3, 0, 0), inplace = True)
```

```python
final['Income_Verified'].unique()
```

```python
final['Income_Verified'].replace(('not verified', 'VERIFIED - income'), (0, 1), inplace = True)
```

```python
final['Home_Owner'].fillna(final['Home_Owner'].mode()[0], inplace=True)
final['Length_Employed'].fillna(final['Length_Employed'].mode()[0], inplace=True)
```

```python
final['Annual_Income'].fillna(final['Annual_Income'].mean(), inplace=True)
final['Months_Since_Deliquency'].fillna(0, inplace=True)
```

```python
final.isnull().sum()
```

```python
a = pd.get_dummies(final[['Gender', 'Purpose_Of_Loan']])
a.head()
```

```python
final.drop(['Gender', 'Income_Verified', 'Home_Owner', 'Purpose_Of_Loan'], axis = 1, inplace = True)
final = pd.concat([final,a],axis = 1, sort = False)
```

```python
final['Annual_Income'] = np.log(final['Annual_Income'])
final['Annual_Income'].hist(bins=20)
```

```python
MIN_MAX_COL = col = ['Loan_Amount_Requested', 'Length_Employed', 'Annual_Income', 'Debt_To_Income', 'Inquiries_Last_6Mo',\
                     'Months_Since_Deliquency', 'Number_Open_Accounts', 'Total_Accounts']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
final[MIN_MAX_COL] = scaler.fit_transform(final[MIN_MAX_COL])
```

```python
train = final[final['label'] == 'train']
test = final[final['label'] == 'test']
```

```python
train.drop('label', axis=1,inplace=True)
test.drop('label', axis=1,inplace=True)
```

```python
train.isnull().sum()
```

```python
y = train['Interest_Rate']
X = train.drop(['Interest_Rate', 'Loan_ID'], axis=1)
ID = test['Loan_ID']
test = test.drop(['Interest_Rate', 'Loan_ID'], axis=1)
```

```python
i=1
kf = StratifiedKFold(n_splits=5,random_state=1,shuffle=True)
for train_index, test_index in kf.split(X_train, y_train):
    print('\n{} of kfold {}'.format(i, kf.n_splits))
#     print(train_index, test_index)
    xtr,xvl = X.loc[train_index],X.loc[test_index]
#     xtr.drop(['index'], axis = 1, inplace = True)
#     print(xtr)
    ytr,yvl = y[train_index],y[test_index]

    model = LogisticRegression(random_state=1)

    model.fit(xtr, ytr)
    pred_test = model.predict(xvl)
    score = accuracy_score(yvl, pred_test)
    print('accuracy_score',score)
    i+=1
# pred = model.predict_proba(xvl)[:,1]
```
